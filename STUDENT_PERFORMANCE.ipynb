```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
students = []
names = []
grades = []

for i in range(10):
    name = input(f"Enter name {i+1}: ")
    names.append(name)

    mark = int(input(f"Enter mark for {name}: "))

    if mark >= 90:
        grade = 'A'
    elif mark >= 80:
        grade = 'B'
    elif mark >= 70:
        grade = 'C'
    elif mark >= 60:
        grade = 'D'
    else:
        grade = 'F'

    grades.append(grade)
    students.append([name, mark, grade])

for student in students:
    print(f"Name: {student[0]}, Marks: {student[1]}, Grade: {student[2]}")
```

```python
names = [student[0] for student in students]
marks = [student[1] for student in students]


plt.figure(figsize=(8, 5))
plt.bar(names, marks, color='pink')

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students' Marks")
plt.ylim(0, 100)
plt.show()
```
